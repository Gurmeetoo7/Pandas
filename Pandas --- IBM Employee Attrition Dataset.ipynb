## Selecting data with `loc[]`(Indexing)


Official guide Link -- https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#indexing


Object Type  ====   Indexers

Series      ===     s.loc[indexer]

DataFrame  ===      df.loc[row_indexer,column_indexer]

```python
row = df.loc[17]

print(type(row))
row.head()
```

### select many rows with `loc[start:finish]`

Pass a slice object of integer row labels to select more than one row.
Notice that both values of the slice object are included in the response, unlike when using `list`.

```python
rows = df.loc[10:13]

print(type(rows))
rows
```

```python
#selecting single employee
age_of_employee_12 = df.loc[12, 'Age']

print(type(age_of_employee_12))
age_of_employee_12
```

```python
age_of_employee_17 = df.loc[17, 'Age']

print(type(age_of_employee_17))
age_of_employee_17
```

#### selecting multiple rows, and multiple columns using `loc[rows, columns]`

now we can use either a single value, a list of values, or a slice object for either the rows or the columns parameters.

```python
df.loc[25:27, 'Age':'DailyRate']
```

```python
df.loc[[13, 70, 12, 2], ['Age', 'DailyRate', 'HourlyRate']]
```

### Selecting rows with a condition


```python
over_30 = df.loc[df.Age > 30]
type(over_30)
```

```python
over_30.head()
```

I’m interested in the BusinessTrvel of the Employees older than 35 years.



```python
business_30=df.loc[df["Age"]>35,["BusinessTravel","Age"]]
business_30.head()
```

I’m interested in rows 10 till 25 and columns 3 to 5.



```python
df.iloc[10:15,3:8]
```

REMEMBER
--When selecting subsets of data, square brackets [] are used.

--Inside these brackets, you can use a single column/row label, a list of column/row labels, a slice   of labels, a conditional expression or a colon.

--Select specific rows and/or columns using loc when using the row and column names

--Select specific rows and/or columns using iloc when using the positions in the table

--You can assign new values to a selection based on loc/iloc
